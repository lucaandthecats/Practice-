```python
# Create Labels for a Plot

# With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

import numpy as np
import matplotlib.pyplot as plt

```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y, color = 'r')
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burn')
plt.show()
```

```python
# Create a Title for a Plot

# To make your report look more organized and professional, you can create a title using the title function.

# e.g add a title to the previous plot

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title('Pulse and calories data')
plt.plot(x, y, color = 'r')
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burn')
plt.show()
```

```python
# you can further personalize your chart using the fontdict parameter in xlabel(), ylabel(), and title()
# to set font properties for the title and labels. It is not strictly necessary, but it can improve the look of your reports

# e.g: for the same chart, set font properties for the title and labels:

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'arial','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.show()

```

```python
# Display Multiple Plots

# You can display multiple plot in one figure by using the subplots() function

# e.g: Draw 2 plots

#plot1

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)

plt.plot(x, y)

#plot2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)

plt.plot(x,y)

plt.show()
```

```python
# the subplot() function takes three argument. The first two arguments represents rows and columns
# while the third one represents the index. One row and two columns, means the plots will be displayed side by side

# plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

# plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.

# So, if we want a figure with 2 rows an 1 column (meaning that the two plots will be displayed 
#on top of each other instead of side-by-side), we can write the syntax like this:

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 1, 1)

plt.plot(x, y)

#plot2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 1, 2)

plt.plot(x,y)

plt.show()
```

```python
# You can add a title to each plot with the title() function:

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Sales')

#plot2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title('Revenue')

plt.show()
```

```python
# You can add a title to the entire figure with the suptitle() function:

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Sales')

#plot2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title('Revenue')

plt.suptitle('Financial data')
plt.show()
```

```python
# Creating Scatter Plots

# You can use the scatter() function to draw and create scatter plots.
# the scatter function will plot one dot for each observation. It will need two arrays, of the same lenght, 
# one for the x-axis and one for the y-axis.

# e.g:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)
plt.show()
```

```python
'''The observation in the example above is the result of 13 cars passing by.

The X-axis shows how old the car is.

The Y-axis shows the speed of the car when it passes.

Are there any relationships between the observations?

It seems that the newer the car, the faster it drives, but that could be a coincidence, after all we only registered 13 cars.'''

# Compare Plots
#So, it seems there is a relationship between the speed of a car and its age. But what if we plot the observations 
# from another day as well? Will the scatter plot tell us something else?

#Draw two plots on the same figure:

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)


```

```python
# You can set your own color for each scatter plot with the color or the c argument:

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'r')

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = 'g')

plt.show()

```

```python
# You can adjust the transparency of the dots with the alpha argument.

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'r', alpha = 1)
plt.xlabel('Age of the car')
plt.ylabel('Speed')

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = 'g', alpha = 0.1)

plt.suptitle('Daily observations of passing cars', size = 15)
plt.show()
```

```python
# You can use the bar() function to draw bar graphs:

# e.g: Draw 4 bars:

import numpy as np
import matplotlib.pyplot as plt

```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
# The bar() function takes arguments that describes the layout of the bars.

# The categories and their values represented by the first and second argument as arrays.

x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)

```

```python
# If you want the bars to be displayed horizontally instead of vertically, use the barh() function:

# e.g: Draw 4 horizontal bars:

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
# The bar() and barh() takes the keyword argument color to set the color of the bars:

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = 'g')
plt.show()
```

```python
# You can use any of the 140 supported color names.

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = 'lime')
plt.show()
```

```python
# The bar() takes the keyword argument width to set the width of the bars:

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.2, color = 'r')
plt.show()

# The default width value is 0.8

```

```python
# The barh() takes the keyword argument height to set the height of the bars:

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()


```

```python
# Histogram

# Histograms are graphs showing frequency distributions, meaning a number of observations within a time inverval

# Histograms can be created with the hist() function

# The hist() function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

# For simplicity we use NumPy to randomly generate an array with 250 values,
# where the values will concentrate around 170, and the standard deviation is 10.

x = np.random.normal(170, 10, 250) # 250 values, mean of 170 and standard deviation of 10

print(x)
```

```python
# The hist() function will read the array and produce a histogram:

plt.hist(x)
plt.show()
```

```python
# Creating Pie Charts

# Pie Charts can be created with the pie() function

y = np.array([60, 10, 20, 10])

plt.pie(y)
plt.show()
# By default the plotting of the first wedge starts from the x-axis and move counterclockwise:

```

```python
# you can add labels to the pie chart using the label() parameter
# The label parameter must be an array with one label for each wedge:

y = np.array([45, 30, 25, 7, 3])

labels = ['Toyota', 'Ford', 'BMW', 'Audi', 'Ferrari']

plt.pie(y, labels = labels)
plt.title('Brands distribution')
plt.show()
```

```python
# if you want one of the elements to standout, you can use the explode() function.

y = np.array([45, 30, 25, 7, 3])

brands = ['Toyota', 'Ford', 'BMW', 'Audi', 'Ferrari']
separation = [0.0, 0, 0.0, 0.3, 0.0]

plt.pie(y, labels = brands, explode = separation)
plt.show
```

```python
# You can add a shadow to the pie chart by setting the shadows parameter to True.


y = np.array([45, 30, 25, 7, 3])

brands = ['Toyota', 'Ford', 'BMW', 'Audi', 'Ferrari']
separation = [0.0, 0, 0.0, 0.3, 0.0]

plt.pie(y, labels = brands, explode = separation, shadow  = True)
plt.show
```

```python
# You can set the color of each wedge with the colors parameter.

# If you decide to specify the colors, you must input an array with one value for each wedge

y = np.array([45, 30, 25, 7, 3])

brands = ['Toyota', 'Ford', 'BMW', 'Audi', 'Ferrari']
colors = ['blue', 'yellow', 'green', 'red']

plt.pie(y, labels = brands, colors = colors)
plt.show
```

```python
# To add a list of explanation for each wedge, use the legend() function:

brands = ['Toyota', 'Ford', 'BMW', 'Audi', 'Ferrari']
colors = ['blue', 'yellow', 'green', 'red']

plt.pie(y, labels = brands, colors = colors)
plt.legend()
plt.show


```

```python
# To add a header to the legend, add the title parameter to the legend function.

brands = ['Toyota', 'Ford', 'BMW', 'Audi', 'Ferrari']
colors = ['blue', 'yellow', 'green', 'red']

plt.pie(y, labels = brands, colors = colors)
plt.legend(title = 'Car brands')
plt.show
```
